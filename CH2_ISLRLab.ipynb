## Graphics

```
x = rnorm(100)
y = rnorm(100)
plot(x,y)
plot(x,y,xlab="this is the x-axis", ylab="this is the y-axis", main="Plot of X vs Y")
```

```
# Creating sequences
seq(1,10)
seq(0,1,length=10)

# Shorthand
1:10
```

```
# Contour plot
# Args:
#   Vector x (first dimension)
#   Vector y (second dimension)
#   Matrix whose elements correspond to the z value (third dimension)

x = 1:10 
y = x
f = outer(x, y, function(x, y) cos(y) / (1 + x ^ 2))

contour(x, y, f)

contour(x, y, f, nlevels=45, add=T)

fa = (f-t(f)) / 2
contour(x, y, fa, nlevels=15)
```

```
# Heatmap
image(x, y, fa)
```

```
# Three Dimensional Plot
persp(x, y, fa)
persp(x, y, fa, theta=30)
persp(x, y, fa, theta=30, phi=20)
persp(x, y, fa, theta=30, phi=70)
persp(x, y, fa, theta=30, phi=40)
```

## Additional Graphical and Numerical Summaries

```
# Use $ to refer to table features
plot(Auto$cylinders, Auto$mpg)
```

```
# Alternative method
attach(Auto)
plot(cylinders, mpg)
```

```
# Convert quantitative variable into a qualitative variable
cylinders = as.factor(cylinders)
```

```
# If the x-axis is categoril, plot will produce boxplots by default
plot(cylinders, mpg)
plot(cylinders, mpg, col="red") # with colors
plot(cylinders, mpg, col="red", varwidth=T) # variable width for boxes
plot(cylinders, mpg, col="red", varwidth=T, horizontal=T) # horizontal orientation
plot(cylinders, mpg, col="red", varwidth=T, xlab="cylinders", ylab="mpg") # with label
```

```
# Histogram
hist(mpg)
hist(mpg, col=2) # 2 == "red"
hist(mpg, col=2, breaks=15) # change breaks length
```
