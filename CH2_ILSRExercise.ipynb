## College Dataset
Prepare the data

```
# Read the csv file
college = read.csv("/content/College.csv");

# Set university as row name
rownames(college) = college[,1]

# Remove the column for university name
college = college[,-1]

college
```

```
# Show numerical summary of the variables in the dataset
summary(college)
```

```
# Convert private to boolean
college$Private = as.factor(college$Private)
summary(college)
```

```
# Show a scatterplot matrix
pairs(college[,1:10])
```

```
# Show plot of Outstate and Private
plot(college$Private, college$Outstate)
```

```
# Create the Elite variable filled with No
Elite = rep("No", nrow(college))

# Set the value to Yes is Top10Prev > 50
Elite[college$Top10perc > 50] = "Yes"

# Convert to numeric
Elite = as.factor(Elite)

# Add Elite to college
college = data.frame(college, Elite)

# Show a summary of the Elite variable
summary(college$Elite)
```

```
# Plot Outstate and Elite
plot(college$Elite, college$Outstate)
```

```
# Show histogram of Expend
hist(college$Expend, col = 2, breaks = 20)
```

```
# Show histogram of Outstate
hist(college$Outstate, col = 2)
```

## Auto Dataset
Prepare the data

```
# Load the dataset
auto = read.csv("/content/Auto.csv", header=T, na.strings="?")

# Drop rows with missing values
auto = na.omit(auto)

# Show head
head(auto)
```

```
# See the types of the variables
sapply(auto, typeof)

# Categorical: name
# Numeric: mpg cylinders displacement horsepower weight acceleration year origin
```

```
# Show the range of each numerical variables
sapply(auto[,1:8], range)
```

```
# Show the mean of each numerical variables
sapply(auto[,1:8], mean)
```

```
# Show the mean of each numerical variables
sapply(auto[,1:8], sd)
```

```
# Remove the 10th-85th row
autoReduced = auto[-c(10:85),]
dim(autoReduced)

# Find the range, mean, and sd with summary
summary(autoReduced[,1:8])
```

```
# Show a scatter plot of all variables
pairs(auto[,1:8])
```
